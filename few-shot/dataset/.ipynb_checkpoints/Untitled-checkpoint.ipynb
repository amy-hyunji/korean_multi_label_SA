```python
import pandas as pd
import numpy as np
import os
import random
```

```python
path = "./after_cleansing/final_after_cleansing_no_neutral.csv"
save_path = "./episodes"

_csv = pd.read_csv(path)
_len = int(len(_csv.index))
sentence = _csv['sentence']
label = _csv['label']
```

```python
happy_list = []
angry_list = []
sad_list = []
surprised_list = []
```

```python
for i in range(_len):
    if (label[i] == 1):
        happy_list.append(sentence[i])
    elif (label[i] == 2):
        sad_list.append(sentence[i])
    elif (label[i] == 3):
        angry_list.append(sentence[i])
    elif (label[i] == 4):
        surprised_list.append(sentence[i])
    elif (label[i] == 5):
        continue
    else:
        print("something wrong with label : {}".format(label[i]))
```

```python
n_episodes = 101
n_shots = 10
n_query = 5
n_classes = 4
t_episodes = 500


for epi in range(t_episodes):
    if not (os.path.exists(os.path.join(save_path, str(epi)))):
        os.mkdir(os.path.join(save_path, str(epi)))
        
    happy_list = np.unique(np.array(happy_list))
    sad_list = np.unique(np.array(sad_list))
    angry_list = np.unique(np.array(angry_list))
    surprised_list = np.unique(np.array(surprised_list))

    t_happy_list = np.random.choice(happy_list, 1560, replace = False)
    t_sad_list = np.random.choice(sad_list, 1560, replace = False)
    t_angry_list = np.random.choice(angry_list, 1560, replace = False)
    t_surprised_list = np.random.choice(surprised_list, 1560, replace = False)

    train_happy = t_happy_list[:1040].tolist()
    train_sad = t_happy_list[:1040].tolist()
    train_angry = t_angry_list[:1040].tolist()
    train_surprised = t_surprised_list[:1040].tolist()

    test_happy = np.setdiff1d(t_happy_list, train_happy).tolist()
    test_sad = np.setdiff1d(t_sad_list, train_sad).tolist()
    test_angry = np.setdiff1d(t_angry_list, train_angry).tolist()
    test_surprised = np.setdiff1d(t_surprised_list, train_surprised).tolist()

    train_dict = {0: train_happy, 1: train_sad, 2: train_angry, 3: train_surprised}
    test_dict = {0: test_happy, 1: test_sad, 2: test_angry, 3: test_surprised}    
    
    for i in range(n_episodes):
        
        train_sentence = []
        train_label = []
        test_sentence = []
        test_label = []

        t_train_sentence = []
        t_train_label = []
        t_test_sentence = []
        t_test_label = []
        
        train_set = []
        test_set = []
        t_train_set = []
        t_test_set = []
        
        if not (os.path.exists(os.path.join(save_path, str(epi), str(i)))):
            os.mkdir(os.path.join(save_path, str(epi), str(i)))
        if not (os.path.exists(os.path.join(save_path, str(epi), "test"))):
            os.mkdir(os.path.join(save_path, str(epi), "test"))
        for j in range(n_classes):
            _train = train_dict[j]
            _test = test_dict[j]
            for k in range(n_shots):
                if (i == n_episodes-1):
                    t_train_set.append([_train[k+i*n_shots], j])
                else:
                    train_set.append([_train[k+i*n_shots], j])
            for q in range(n_query):
                if (i == n_episodes-1):
                    t_test_set.append([_test[q+i*n_query], j])
                else:
                    test_set.append([_test[q+i*n_query], j])
                    
        random.shuffle(t_train_set)
        random.shuffle(train_set)
        random.shuffle(t_test_set)
        random.shuffle(test_set)
        
        for elem in t_train_set:
            t_train_sentence.append(elem[0])
            t_train_label.append(elem[1])
        for elem in train_set:
            train_sentence.append(elem[0])
            train_label.append(elem[1])
        for elem in t_test_set:
            t_test_sentence.append(elem[0])
            t_test_label.append(elem[1])
        for elem in test_set:
            test_sentence.append(elem[0])
            test_label.append(elem[1])
                    
        pd_train = pd.DataFrame({'sentence': train_sentence, 'label': train_label})
        pd_train.to_csv("./episodes/{}/{}/train.csv".format(epi, i))
        pd_test = pd.DataFrame({'sentence': test_sentence, 'label': test_label})
        pd_test.to_csv("./episodes/{}/{}/test.csv".format(epi, i))
        pd_t_train = pd.DataFrame({'sentence': t_train_sentence, 'label': t_train_label})
        pd_t_train.to_csv("./episodes/{}/test/train.csv".format(epi))
        pd_t_test = pd.DataFrame({'sentence': t_test_sentence, 'label': t_test_label})
        pd_t_test.to_csv('./episodes/{}/test/test.csv'.format(epi))
        
```
